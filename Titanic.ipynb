```python
train = train.drop('Agemode12', axis = 1)
test = test.drop('Agemode12', axis = 1)
```

```python
train[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)

```

```python
train['Embarked'] = train['Embarked'].map( {'S': 0.3, 'C': 0.5, 'Q': 0.4} ).astype(float)
test['Embarked'] = test['Embarked'].map( {'S': 0.3, 'C': 0.5, 'Q': 0.4} ).astype(float)
```

```python
train.loc[ train['Age'] <= 16, 'Age'] = 0
train.loc[(train['Age'] > 16) & (train['Age'] <= 32), 'Age'] = 1
train.loc[(train['Age'] > 32) & (train['Age'] <= 48), 'Age'] = 2
train.loc[(train['Age'] > 48) & (train['Age'] <= 64), 'Age'] = 3
train.loc[ train['Age'] > 64, 'Age'] = 4

test.loc[ test['Age'] <= 16, 'Age'] = 0
test.loc[(test['Age'] > 16) & (test['Age'] <= 32), 'Age'] = 1
test.loc[(test['Age'] > 32) & (test['Age'] <= 48), 'Age'] = 2
test.loc[(test['Age'] > 48) & (test['Age'] <= 64), 'Age'] = 3
test.loc[ test['Age'] > 64, 'Age'] = 4

```

```python
name_length = [len(name) for name in train['Name']]
avg = sum(name_length)/len(name_length)

train.loc[np.array(name_length) <= avg,"Name"] = "short"
train.loc[np.array(name_length) > avg,"Name"] = "long"



train[['Name', 'Survived']].groupby(['Name'], as_index=False).mean().sort_values(by='Survived', ascending=False)

```

```python



test.loc[np.array(name_length) <= avg,"Name"] = "short"
test.loc[np.array(name_length) > avg,"Name"] = "long"

Name_dummies_train = pd.get_dummies(train['Name'])
Name_dummies_test = pd.get_dummies(test['Name'])

train = train.drop('Name', axis= 1)
test = test.drop('Name' , axis= 1)
                 
train = train.join(Name_dummies_train)
test = test.join(Name_dummies_train)
```
